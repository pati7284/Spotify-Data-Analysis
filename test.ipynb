```python
import pandas as pd
import re

charts = pd.read_csv(
    "spotify_charts.csv",
    usecols=["title", "artist", "url"],
    nrows=20
)

def extract_track_id(url):
    match = re.search(r"track/([A-Za-z0-9]+)", str(url))
    return match.group(1) if match else None

charts["track_id"] = charts["url"].apply(extract_track_id)
charts.head(10)
```

```python
songs = pd.read_csv(
    "spotify_songs.csv",
    usecols=["track_id", "track_name", "track_artist"]
)

test_merge = charts.merge(songs, on="track_id", how="left")
test_merge[["title", "artist", "track_id", "track_name", "track_artist"]].head(10)
```
