```python
import pandas as pd
import re
import os

charts_file = "spotify_charts.csv"
songs_file = "spotify_songs.csv"
output_file = "spotify_merged.csv"
```

```python
# =========================
# 2. Wczytanie mniejszego pliku songs
# =========================
songs_cols = [
    "track_id", "track_name", "track_artist", "track_popularity",
    "playlist_genre", "playlist_subgenre",
    "danceability", "energy", "key", "loudness", "mode",
    "speechiness", "acousticness", "instrumentalness",
    "liveness", "valence", "tempo", "duration_ms"
]

songs = pd.read_csv(songs_file, usecols=songs_cols)

# Usunięcie duplikatów po track_id
songs = songs.drop_duplicates(subset="track_id")

print("Wczytano songs:", songs.shape)

# =========================
# 3. Funkcja do wyciągania track_id z URL
# =========================
def extract_track_id(url):
    if pd.isna(url):
        return None
    
    match = re.search(r"track/([A-Za-z0-9]+)", str(url))
    if match:
        return match.group(1)
    return None
# =========================
# 4. Kolumny z dużego pliku charts
# =========================
charts_cols = [
    "title", "rank", "date", "artist", "url",
    "region", "chart", "trend", "streams"
]

# Jeśli plik wynikowy już istnieje, usuń go
if os.path.exists(output_file):
    os.remove(output_file)

# =========================
# 5. Czytanie charts partiami
# =========================
chunk_size = 100_000
first_chunk = True
chunk_number = 0

for chunk in pd.read_csv(charts_file, usecols=charts_cols, chunksize=chunk_size):
    chunk_number += 1
    print(f"Przetwarzanie chunk {chunk_number}...")

    # Wyciągnięcie track_id z url
    chunk["track_id"] = chunk["url"].apply(extract_track_id)

    # Merge po track_id
    merged_chunk = chunk.merge(songs, on="track_id", how="left")

    # Zapis do pliku wynikowego
    merged_chunk.to_csv(
        output_file,
        mode="w" if first_chunk else "a",
        index=False,
        header=first_chunk
    )

    first_chunk = False

print("Gotowe. Zapisano do:", output_file)

merged_sample = pd.read_csv("spotify_merged.csv", nrows=10)
print(merged_sample.head())
```
