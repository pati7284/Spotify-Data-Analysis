```python
import pandas as pd

merged_data = pd.read_csv("spotify_merged.csv")

clean_spotify_data = merged_data.dropna(subset=["loudness"]).copy()
print("Shape przed czyszczeniem:", merged_data.shape)
print("Shape po usunięciu NA z loudness:", clean_spotify_data.shape)
```

```python
columns = clean_spotify_data.columns.tolist()
print("Kolumny w zbiorze danych:", columns)

for column in columns:
    missing_count = clean_spotify_data[column].isna().sum()
    print(f"Liczba brakujących wartości w kolumnie '{column}': {missing_count}")

clean_spotify_data = clean_spotify_data.drop(columns=["streams"])
```

```python
clean_spotify_data.to_csv("clean_spotify_data.csv", index=False)
```
